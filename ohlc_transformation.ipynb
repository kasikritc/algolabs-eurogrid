```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('2024.2.21EURUSD-TICK-No Session.csv')
df
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%Y%m%d %H:%M:%S.%f')
df
```

```python
def resample_tick_data(df, timeframe):
    modified_df = df.copy(deep=True)
    modified_df.set_index('DateTime', inplace=True)

    # Calculate the 'mid' price from Bid and Ask
    modified_df['Mid'] = (modified_df['Bid'] + modified_df['Ask']) / 2

    # Resample and aggregate data
    ohlc = modified_df['Mid'].resample(timeframe).ohlc()
    volume = modified_df['Volume'].resample(timeframe).sum()

    # Create a new DataFrame with both OHLC and Volume
    ohlc_df = pd.DataFrame(index=ohlc.index)
    ohlc_df['Open'] = ohlc['open']
    ohlc_df['High'] = ohlc['high']
    ohlc_df['Low'] = ohlc['low']
    ohlc_df['Close'] = ohlc['close']
    ohlc_df['Volume'] = volume

    return ohlc_df
```

```python
ohlc_df = resample_tick_data(df, '1MIN')
ohlc_df
```

```python
ohlc_df.to_csv('1MIN_EURUSD_OHLC.csv')
```
